# Нормализация датасета приложений в Google Play

## Обработка бинарных признаков

```python
gp = gp_data.copy()
print(gp['Type'].unique())
```

```python
gp["Is Paid"] = np.where(gp_data["Type"] == "Paid", 1, 0)
```

```python
gp.drop('Type', axis=1, inplace=True)
```

```python
gp.head(10)
```

## Обработка номинальных признаков

```python
print(gp['Category'].unique())
print(gp["Category"].value_counts().sort_values(ascending=False)[:2])
```

Две самые популярные категории: FAMILY и GAME. В нормализованной таблице будут три колонки, (указанные выше плюс Other Categories).

```python
dummies = pd.get_dummies(gp["Category"], prefix="Category", prefix_sep="_")

dummies = dummies[["Category_FAMILY", "Category_GAME"]]
dummies.loc[
    (dummies['Category_FAMILY'] == 0) &
    (dummies['Category_GAME'] == 0),
    'Category_Other'
] = True
dummies['Category_Other'].fillna(False, inplace=True)
dummies = dummies.astype(int)

gp = pd.concat([gp, dummies], axis=1)
gp.drop('Category', axis=1, inplace=True)
gp.head(10)
```

```python
gp.info()
```

```python
gp1 = gp.copy()
print(gp1['Content Rating'].unique())
print(gp1['Content Rating'].value_counts().sort_values(ascending=False)[:2])
```

```python
dummies = pd.get_dummies(gp1["Content Rating"], prefix="ContentRating", prefix_sep="_")
dummies_main = dummies[['ContentRating_Everyone', 'ContentRating_Teen']]
dummies_main.loc[
    (dummies_main['ContentRating_Everyone'] == False) &
    (dummies_main['ContentRating_Teen'] == False),
    'ContentRating_Other'
] = True
dummies_main['ContentRating_Other'].fillna(False, inplace=True)
dummies_main = dummies_main.astype(int)

gp1 = pd.concat([gp1, dummies_main], axis=1)
gp1.drop('Content Rating', axis=1, inplace=True)
gp1.head(10)
```

```python
gp2 = gp1.copy()
print(gp1['Genres'].unique())
print(gp1['Genres'].value_counts().sort_values(ascending=False)[:2])
```

```python
dummies = pd.get_dummies(gp2["Genres"], prefix="Genres", prefix_sep="_")
dummies_main = dummies[['Genres_Tools', 'Genres_Entertainment']]
dummies_main.loc[
    (dummies_main['Genres_Tools'] == False) &
    (dummies_main['Genres_Entertainment'] == False),
    'Genres_Other'
] = True
dummies_main['Genres_Other'].fillna(False, inplace=True)
dummies_main = dummies_main.astype(int)

gp2 = pd.concat([gp2, dummies_main], axis=1)
gp2.drop('Genres', axis=1, inplace=True)
```

```python
gp2.head(10)
```

### Порядковые признаки

```python
gp3 = gp2.copy()

max_rating = gp3['Rating'].max()
gp3['Rating'] = gp3['Rating'] / max_rating
```

```python
gp3.head(10)
```

### Количественные признаки

```python
def value_to_int(s):
    if 'k' in s:
        return int(float(s.replace('k', '')) * 1000)
    elif 'M' in s:
        return int(float(s.replace('M', '')) * 1000000)
    else:
        return int(s)
```

```python
gp4 = gp3.copy()

gp4['Reviews'] = gp4['Reviews'].apply(value_to_int) # normalizing the data
reviews = gp4['Reviews']
reviews_mean = reviews.sort_values(ascending=False).mean()

gp4['Reviews'] = gp4['Reviews'].apply(lambda x: 0 if x <= reviews_mean else 1)
```

```python
gp4.head(10)
```

```python
gp5 = gp4.copy()

gp5['Price'] = gp5['Price'].str.replace('Everyone', '0')
gp5['Price'] = gp5['Price'].str.replace('$', '')

gp5['Price'] = gp5['Price'].astype(float)

mean_price = gp5['Price'].mean()

gp5['Price'] = gp5['Price'].apply(lambda x: 0 if x <= mean_price else 1)

gp5.head(10)
```

```python
gp6 = gp5.copy()

gp6['Installs'] = gp6['Installs'].str.replace('+', '')
gp6['Installs'] = gp6['Installs'].str.replace(',', '')
gp6['Installs'] = gp6['Installs'].str.replace('Free', '0')

gp6['Installs'] = gp6['Installs'].astype(int)

mean_installs = gp6['Installs'].mean()

gp6['Installs'] = gp6['Installs'].apply(lambda x: 0 if x <= mean_installs else 1)

gp6.head(10)
```

```python
gp7 = gp6.copy()

gp7['Size'] = gp7['Size'].str.replace('Varies with device', '0M')
gp7['Size'] = gp7['Size'].str.replace(',', '')
gp7['Size'] = gp7['Size'].str.replace('+', '')

gp7['Size'] = gp7['Size'].apply(value_to_int)

mean_size = gp7['Size'].mean()

gp7['Size'] = gp7['Size'].apply(lambda x: 0 if x <= mean_size else 1)

gp7.head(10)
```

```python
gp7.drop(['Current Ver', 'Android Ver', 'Last Updated'], axis=1, inplace=True)
```

```python
gp7.head(10)
```
